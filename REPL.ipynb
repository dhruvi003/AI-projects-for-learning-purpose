```python
## let's try to pass the entire message history
# Re-initialize messages at the start of the cell to ensure a clean slate
messages = [
    {"role": "system", "content": "You are a helpful assistant."}
]

while True:
    user_input = input("Enter: ")

    # Fix 1: user_msg should be a dictionary, not a list containing a dictionary
    # Fix 2: Change 'context' to 'content' for the user message
    user_msg = {"role":"user", "content": user_input}

    messages.append(user_msg)
    if user_input.lower() == "quit":
        break

    response = client.responses.create(
        model="llama-3.3-70b-versatile",
        input=messages,
    )

    ai_reply = response.output_text
    messages.append({"role": "assistant", "content": ai_reply})

    print(f"AI: {ai_reply}\n")
```
