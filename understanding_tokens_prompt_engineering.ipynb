```python
def count_token(text: str):
    encoder = tiktoken.get_encoding("cl100k_base")
    encoding = encoder.encode(text)
    return len(encoding)
```

```python
def output_token_message(text):
  r = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": text}],
        temperature=0 )
  return r.usage.completion_tokens
```

```python
def calculate_cost(txt):
    # let us calculate as below
    """
    Llama 3.1 8B Instant 128k
    840 TPS
    input_token_cost = $0.05 (20M / $1)*
    output_token_cost = $0.08 (12.5M / $1)*
    """
    input_token_count = count_token(txt)
    output_token_count = output_token_message(txt)

    input_cost =( input_token_count/1_000_000) * 0.05
    output_cost = (output_token_count/1_000_000) * 0.08

    total_cost = (input_cost + output_cost)

    return total_cost
```

```python
print(calculate_cost("Summarise this 10 page document: " + "word " * 2000))
```
