```python
import torch
```

```python
from transformers import pipeline
```

```python
text_translator = pipeline("translation", model="facebook/nllb-200-distilled-600M", dtype=torch.bfloat16)
```

```python
text = "Hello, how are you?"
```

```python
# here https://github.com/facebookresearch/fairseq/tree/main/fairseq
```

```python
translation = text_translator(text, src_lang="eng_Latn", tgt_lang="deu_Latn") # deu -> german
```

```python
import json
```

```python
file_path = "/content/language.json"
```

```python
def get_language_code(language, json_path=file_path):
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    for item in data:
        if item["language"].lower() == language.lower():
            return item["code"]

    return None
```

```python
get_language_code("English")
```

```python
def translate_text(text, dest_lang):
  # text = "Hi"
  dest_code = get_language_code(dest_lang)
  translation = text_translator(text, src_lang="eng_Latn", tgt_lang=dest_code)
  return translation[0]["translation_text"]
```

```python
import gradio as gr
```

```python
demo = gr.Interface(
    fn = translate_text,
    inputs = [gr.Textbox(label="input text to translate", lines=6), gr.Dropdown(choices = ["German","French","Hindi","Spanish","Japanese"],label="select destination language")],
    outputs = [
        gr.Textbox(label="Translated text", lines=4)
    ],
    title = "multilanguage translator",

)
```

```python
demo.launch()
```
