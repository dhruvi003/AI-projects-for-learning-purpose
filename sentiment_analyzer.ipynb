```python
# Use a pipeline as a high-level helper
from transformers import pipeline
```

```python
import torch
import gradio as gr
```

```python
pipe = pipeline("text-classification", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
model_path = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"
```

```python
pipe("This product is good")
```

```python
a = pipe("wow, it is so good that it makes me bald even though it should not")
```

```python
a[0]['label']
```

```python
def sentiment_analysis(input):
    sentiment = pipe(input)
    return sentiment[0]['label']
```

```python
a = sentiment_analysis("good")
a
```

```python
demo = gr.Interface(
    fn = sentiment_analysis,
    inputs = [gr.Textbox(
        label="Input Text",
        lines=1
    )],
    outputs = [
        gr.Textbox(
            label="Sentiment",
            lines=1
        )
    ],
    title="Sentiment Analysis",
    description="Enter text to analyze sentiment"
)
```

```python
demo.launch()
```

```python
import pandas as pd
```

```python
def create_sentiment_dataframe(file_path):
    # Read excel file
    df = pd.read_excel(file_path)

    # Assuming the review column name is 'review'
    reviews = df['reviews']

    # Get sentiment for each review
    sentiments = reviews.apply(lambda x: sentiment_analysis(x))

    # Create new dataframe with review + sentiment
    result_df = pd.DataFrame({
        "review": reviews,
        "sentiment": sentiments
    })

    return result_df
```
